```python
import pandas as pd
import unicodedata
import re
```

```python
df_pt = pd.read_csv('../../dataset/pt/atas_dataset.csv')
df_en = pd.read_csv('../../dataset/en/atas_copom_ingles.csv')
```

```python
df_pt.head()
```

```python
def remove_accents_and_nonascii(text):
    if not isinstance(text, str):
        return text  # Evita erros se o valor for NaN ou não-string
    text = unicodedata.normalize('NFKD', text)
    text = text.encode('ASCII', 'ignore').decode('ASCII')
    text = re.sub(r'\s+', ' ', text)
    return text.strip()
```

```python
df_pt['texto_ata'] = df_pt['texto_ata'].apply(remove_accents_and_nonascii)
df_en['texto_ata'] = df_en['texto_ata'].apply(remove_accents_and_nonascii)
```

```python
df_pt = df_pt[~(df_pt['data'] // 10000 == 2000)]
df_pt = df_pt[~(df_pt['data'] // 10000 == 2001)]
df_pt.tail()
```

```python
df_en = df_en[~(df_en['data'] // 10000 == 2000)]
df_en = df_en[~(df_en['data'] // 10000 == 2001)]
df_en.tail()
```

```python
df_en.head()
```

```python
df_pt.to_csv('../../dataset/atas_pt_dataset.csv')
df_en.to_csv('../../dataset/atas_en_dataset.csv')
df_pt.to_parquet('../../dataset/atas_pt_dataset.parquet')
df_en.to_parquet('../../dataset/atas_en_dataset.parquet')
```
