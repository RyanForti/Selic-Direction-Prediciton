```python
#Script para ajustar os datasets pt e en para serem usados no fine-tunning apenas com as colunas texto e decisao_seguinte/decision_next
import pandas as pd
```

```python
df_pt = pd.read_csv('../../dataset/atas_pt_dataset.csv')
df_en = pd.read_csv('../../dataset/atas_en_dataset.csv')
```

```python
#Os dados estão ordenados decrescentemente pela data, ou seja, data mais recentes tem índice menor
df_pt["decisao_n+1"] = df_pt["decisao"].shift(1)
df_pt = df_pt.dropna().reset_index(drop=True)

df_pt
```

```python
df_en["decisao_n+1"] = df_en["decisao"].shift(1)
df_en = df_en.dropna().reset_index(drop=True)

df_en
```

```python
df_pt = df_pt.drop(columns=['Unnamed: 0'])
df_pt
```

```python
df_en = df_en.drop(columns=['Unnamed: 0'])
df_en
```

```python
df_pt.to_csv('../../dataset/dataset_pt.csv')
df_en.to_csv('../../dataset/dataset_en.csv')
```
